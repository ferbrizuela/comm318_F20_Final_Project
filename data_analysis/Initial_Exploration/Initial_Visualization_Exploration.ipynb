# Sales Revenue Data Set

```python
#Loading Revenue Data
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
revenue_df = pd.read_csv('../../data/Old_Data/Music_Sales_Revenue_Clean.csv')
```

```python
revenue_df
```

```python
# Create subset for specific years
revenue_2000_df = revenue_df[revenue_df['Year']==2000]
```

```python
revenue_2000_df.plot(x = "Format", y = "Revenue_Value", kind = "bar")
plt.show()
```

```python
revenue_2010_df = revenue_df[revenue_df['Year']==2010]
```

```python
revenue_2010_df.plot(x = "Format", y = "Revenue_Value", kind = "bar")
plt.show()
```

```python
revenue_2019_df = revenue_df[revenue_df['Year']==2019]
```

```python
revenue_2019_df.plot(x = "Format", y = "Revenue_Value", kind = "bar")
plt.show()
```

## Sales Volume Comparison of Formats

```python
volume_df.groupby(['Year','Format'])['Units_Sold'].sum().unstack().plot(figsize=(12,4))
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

# Sales Revenue Comparison of Formats

```python
revenue_df.groupby(['Year','Format'])['Revenue_Value'].sum().unstack().plot(figsize=(12,4))
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
