## Q1. How many different formats were used to consume music in 2000 vs. 2019?

* Steps: 
    1. find the rows that pertain to year `2000` in `volume_df`
        * use `Year` column to find the rows that have 2000 as date. 
    2. create a subset of the year 2000 data named `formats_df`
    3. note the amount of rows, which indicate the number of formats in that year.
    4. repeat with 2019 information. 


```python
#Find the rows that pertain to year 2000 in volume_df
formats_2000 = volume_df['Year']==2000
```

```python
#Create a subset of the year 2000 data named formats_df
volume_df[formats_2000]
```

```python
#Note the amount of rows, which indicate the number of formats in that year.
volume_df[formats_2000].shape
formats_2000_nrows = volume_df[formats_2000].shape[0]

formats_2000_nrows, formats_2000_ncols = volume_df[formats_2000].shape

print(f'There were {formats_2000_nrows} different types of music formats sold in the year 2000')
```

```python
#Find the rows that pertain to year 2019 in volume_df
formats_2019 = volume_df['Year']==2019
```

```python
#Create a subset of the year 2019 data named formats_df
volume_df[formats_2019]
```

```python
#Note the amount of rows, which indicate the number of formats in that year.
volume_df[formats_2019].shape
formats_2019_nrows = volume_df[formats_2019].shape[0]

formats_2019_nrows, formats_2019_ncols = volume_df[formats_2019].shape

print(f'There were {formats_2019_nrows} different types of music formats sold in the year 2019')
```

#### Answer: 
* In 2000 there were 9 different music formats, whereas in 2019 there were 12 different music formats.

#### Observations: 
* Although more information has to be interpreted and there certainly is more to the story, the increase of available musical formats from 2000 to 2019 can be an indicator as to an increase in music sales (in terms of volume) in the past decade. 
* Looking through the subsets in each year, there is an indication that the digitalization of music caused an increase in available musical formats. 

```python
volume_df.groupby('Year')['Format'].nunique()
```

```python
formats_by_year = volume_df.groupby('Year')['Format'].nunique()
formats_by_year.plot(kind='bar')
```

## Q2. The CD is a format that seems to be present throughout the decade. How have CD sales (in volume) changed over time from 2000 to 2019?

* Steps: 
    1. Create a subset, a data frame focusing only on CD sales throughout the years. 
    2. Look at the `Format` and `Value (Actual)` columns to see how sales have changed. 
    3. Plot the data to have a visualization of how this has changed throughout the decade. 

```python
#Create a subset, a data frame focusing only on CD sales throughout the years. 
CD_sales_df = volume_df[volume_df['Format']=='CD']
```

```python
CD_sales_df
```

```python
# Just doing this to see spacing in column names.
CD_sales_df.columns
```

```python
# Look at the `Format` and `Value (Actual)` columns to see how sales have changed.
CD_sales_df[['Format','Value (Actual)']]
```

```python
# Also interested in % of total volume, it's noticeable that CDs used to be the 
# most sold and contributed the most to music sale volume and that has definitely changed.
CD_sales_df[['Format','% of Total Volume']]
```

### I am really interested in plotting % of Total Volume in addition to Value (Actual) but I get an error indicating no numeric data?

#### Answer: 
* It is clear both by the numbers in the data frame and the graph that CD sales have declined since 2000. 

#### Observations: 
* CD sales have been on a steady decline in terms of volume. I am also extremely interested in looking at the percentage of total volume but have not been able to plot it because of a 'no numeric data' error.
