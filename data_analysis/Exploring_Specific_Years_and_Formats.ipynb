```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
music_sales_df = pd.read_csv('../data/Usable_Data/Usable_Music_Sales_Data.csv')
```

#### In this notebook, we will look into specific years and formats. Using the visualizations in the other exploration notebooks, I will look at the years that had the most sales. Additionally, I did some comparisons based on specific formats.

#### First, we will do an exploration in terms of sales volume. Looking at the year 1973 which is the first in the data set, then 1999, 2008, 2014, and 2019. These years indicate major changes in the music industry. 

```python
music_sales_df = music_sales_df.astype({"Year":'str'})
```

```python
volume_df = music_sales_df[music_sales_df['Metric']=='Units']
```

```python
volume_1973_df = volume_df[volume_df['Year']=='1973']
volume_1973_df
```

#### In 1973, there were 5 formats, the most popular was the LP/EP

```python
volume_1999_df = volume_df[volume_df['Year']=='1999']
volume_1999_df
```

#### In 1999, we can see an increase in formats from 5 in 1973 to 9 in 1999. This includes the introduction of the CD in 1983. The most popular format in 1999 was, by far, the CD. 

```python
volume_2008_df = volume_df[volume_df['Year']=='2008']
volume_2008_df
```

```python
volume_2008_df.shape
```

#### In 2008, there were 17 formats (considering that CD and CD single were sold separately) and digital downloads were already available. It is interesting to note that the most popular format is Download Single. When CDs were the most popular format, CD of full albums were more popular than CD singles. This indicates that the way people listened to and purchased music changed. People were no longer constrained to purchasing full albums. Individual songs were now more accessible than ever.

```python
volume_2014_df = volume_df[volume_df['Year']=='2014']
volume_2014_df
```

```python
volume_2014_df.shape
```

#### In 2014, formats decreased in number. Out of the 13 formats, only 3 are analog. Streaming is already present this year. In terms of volume, however, Download  Single continues to be the most popular.

```python
volume_2019_df = volume_df[volume_df['Year']=='2019']
volume_2019_df
```

```python
volume_2019_df.shape
```

####  There are again 13 formats. However, one of the analog formats was replaced for a digital. Download Single continues to be the most popular, as Streaming is not a major format when looking at sales _volume_. 

#### Now we will look at the revenue data.

```python
revenue_df = music_sales_df[music_sales_df['Metric']=='Value (Adjusted)']
```

```python
revenue_1973_df = revenue_df[revenue_df['Year']=='1973']
revenue_1973_df
```

#### In 1973, the most popular format (LP/EP) in terms of sales in volume is also the one that brings in the most revenue.

```python
revenue_1999_df = revenue_df[revenue_df['Year']=='1999']
revenue_1999_df
```

#### The same is true in 1999, CD is the most popular in terms of sales volume and revenue. In 1999 it seems like some Analog formats continue to bring in revenue more than other digital formats. 

```python
revenue_2008_df = revenue_df[revenue_df['Year']=='2008']
revenue_2008_df
```

#### The same is true in 2008. However, this year, most formats bringing in significant revenue are digital.

```python
revenue_2014_df = revenue_df[revenue_df['Year']=='2014']
revenue_2014_df
```

#### In 2014, Streaming formats start to become significant revenue streams. Analog formats are mostly bring in little to no revenue, as well as some physical digital formats.

```python
revenue_2019_df = revenue_df[revenue_df['Year']=='2019']
revenue_2019_df
```

#### In 2019, Paid subscription is the format that contributes the most revenue to the music industry. Download single, which is the one that leads sales in terms of _volume_ contributes even _less_ revenue than the LP/EP format which is analog. 

```python
volume_album_df = volume_df[volume_df["Format"].isin(["CD","Download Album"])]
```

```python
volume_album = volume_album_df.groupby(['Year','Format'])['Actual_Value'].sum()
```

```python
volume_album.unstack().plot(ylabel="Units (Millions)",title="Recorded Album Sales Volumes",
                           color=("green","blue"))
```

```python
volume_single_df = volume_df[volume_df["Format"].isin(["CD Single","Download Single"])]
```

```python
volume_single = volume_single_df.groupby(['Year','Format'])['Actual_Value'].sum()
```

```python
volume_single.unstack().plot(ylabel="Units (Millions)",title="Recorded Single Sales Volumes",
                            color=("green","blue"))
```
