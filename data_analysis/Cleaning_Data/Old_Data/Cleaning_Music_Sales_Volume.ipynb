```python
#Loading Data
import pandas as pd
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
volume_df = pd.read_csv('../../../data/Old_Data/Music_Sales_Volume_Data.csv')
```

## Cleaning Music Sales Volume Data

```python
volume_df.shape
```

```python
volume_df.head(20)
```

```python
volume_df2 = volume_df.rename(columns={'Value (Actual)':'Units_Sold','% of Total Volume': 'Percent_Of_Total_Volume', 'Format Value # (Million)': 'Format_Value_Num_Million',
                                     'Total Value # (Billion)':'Total_Value_Num_Billion','Total Value # (Million)':'Total_Value_Num_Million',
                                     'Total Value For Year':'Total_Value_For_Year'})
```

```python
volume_df2.columns
```

```python
cols_to_keep= ['Format','Metric','Units_Sold','Year','Percent_Of_Total_Volume','Total_Value_Num_Million','Total_Value_Num_Billion','Total_Value_For_Year']
```

```python
cols_to_keep
```

```python
volume_df3=volume_df2[cols_to_keep]
```

```python
volume_df3
```

### Notes:

* Renamed column titles to remove spaces and other signs such as % and #
* Removed columns that were unneeded, some were copies and some were just irrelevant. Others were missing data.
* Units_Sold will be the most relevant column in terms of quantifying sales. 
* Kept 'Total_Value_Num_Million','Total_Value_Num_Billion','Total_Value_or_Year' just for context, but will not be relevant when analyzing this data frame as the Revenue data frame will bring more context into sales in terms of Revenue. 

```python
volume_df3['Percent_Of_Total_Volume']
```

```python
volume_df3['Percent_Of_Total_Volume'] = volume_df3['Percent_Of_Total_Volume'].str.replace('%','')
```

```python
volume_df3['Percent_Of_Total_Volume']
```

```python
volume_df3
```

```python
volume_df3["Percent_Of_Total_Volume"] = volume_df3.Percent_Of_Total_Volume.astype(float)
```

```python
volume_df3['Percent_Of_Total_Volume']
```

## Note:

* Removed '%' from Percent_Of_Total_Volume and turned contents into floats
* This number may be relevant for figuring out the format that is more prevalent within a year. 
    * So for example, in the year 2000 we can see that CD sales were 87.3% of total volume of sales in the year. So that format was the most popular.
